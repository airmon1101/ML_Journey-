```python
def yelp(price):
  if price < 10:
    return '$'
  elif price >= 10 and price < 30:
    return '$$'
  else:
    return '$$$'
```

```python
df['total_bill'].apply(yelp)
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df.sample(20)
```

```python
def quality(total_bill, tip):
  if tip / total_bill > 0.25:
    return 'Generous'
  else:
    return 'Other'
```

```python
df['Quality'] = df.apply(lambda row: quality(row['total_bill'], row['tip']), axis=1)
```
