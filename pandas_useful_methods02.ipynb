```python
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
# Shorts values using 'sort_values()'
df.sort_values('tip', ascending=True) # by default 'ascending=True'
```

```python
# you can pass a list inside this '()'
df.sort_values(['tip', 'size'], ascending=False) # by default 'ascending=True'
```

```python
# Fint the max value index location
df['total_bill'].max() # for return the max value
df['total_bill'].idxmax() # return max value indx location
```

```python
df.iloc[170] # return max value index
```

```python
# OR
df.iloc[df['total_bill'].idxmax()]
```

```python
# Quick correlation check using '.corr()'
# correlation = df.corr()

correlation = df.select_dtypes(include=['number']).corr()
```

```python
correlation
```

```python
# Value count using '.value_counts'
df.value_counts() # Q==f
```

```python
# Check unique value in DataFrame using '.unique()' OR '.nunique()'
df['sex'].unique() # return unique values
```

```python
df['sex'].nunique() # return number of unique value in int
```

```python
df['day'].unique() # return unique values
```

```python
df['day'].nunique() # return number of unique value in int
```

```python
# Replace method
df['sex'].replace('Female', 'F')
```

```python
# For multiple replace
df['sex'].replace(['Female', 'Male'], ['F', 'M'])
```

```python
# Replace using 'Map' method
# Note it's best for large data set's
# And it's easy to read :)
mymap = {'Female':'F', 'Male':'M'}
df['sex'].map(mymap)
```

```python
# '.nlargest()' return's largest "tip" in 'df'
df.nlargest(3, 'tip') # (Row, 'Column') \ (no_of_row_you_want, 'Column_name')
```

```python
# '.nsmallest()' return's smallest "tip" in 'df'
df.nsmallest(3, 'tip') # (Row, 'Column') \ (no_of_row_you_want, 'Column_name')
```
