```python
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
# Columns are feature
# Rows are Instances of data
# This formate is required for 'ML' later on
```

```python
df.head()
```

```python
bool_series = df['total_bill'] > 40
```

```python
df[bool_series]
```

```python
# OR
```

```python
df[df['total_bill'] > 40]
```

```python
bool_series = df['sex'] == "Male"
df[(bool_series)]

# OR

df[(df['sex'] == "Male")]
```

```python
df[df['size'] > 3]
```

```python
# Multiple Conditions
# AND & --> BOTH CONDITION NEED TO BE TRUE
# OR | --> EITHER CONDITION IS TRUE
```

```python
df[(df['total_bill'] > 25) & (df['sex'] == "Female")]
```

```python
df[(df['total_bill'] > 25) | (df['sex'] == "Male")]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat') | (df['day'] == 'Thur')]
```
